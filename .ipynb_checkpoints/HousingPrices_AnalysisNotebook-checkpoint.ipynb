```python
import numpy as np # library to handle data in a vectorized manner
import pandas as pd # library for data analsysis
from numerize.numerize import numerize as nz # Easy number formatting
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator,StrMethodFormatter, FuncFormatter, AutoMinorLocator)
```

```python
# Load All Homes CSV file
Neighborhood_Zhvi_AllHomes = pd.read_csv('Data/Neighborhood_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv',index_col=[5,7,8,6,2])
Neighborhood_Zhvi_2bedroom = pd.read_csv('Data/Neighborhood_zhvi_bdrmcnt_2_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv',index_col=[5,7,8,6,2])
Neighborhood_Zhvi_3bedroom = pd.read_csv('Data/Neighborhood_zhvi_bdrmcnt_3_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv',index_col=[5,7,8,6,2])
```

```python
Neighborhood_Zhvi_AllHomes.head()
```

```python
lastcol = Neighborhood_Zhvi_AllHomes.columns[-1] #Define last column to use for current prices
LastPriceDF = Neighborhood_Zhvi_AllHomes.iloc[:,[-1]].rename(columns={lastcol:'AllHomes'})
LastPriceDF.join(
    Neighborhood_Zhvi_2bedroom.iloc[:,[-1]].rename(columns={lastcol:'2bed'})).join(
    Neighborhood_Zhvi_3bedroom.iloc[:,[-1]].rename(columns={lastcol:'3bed'})
)
```

```python
Top5States = LastPriceDF.groupby(level='State').mean().sort_values(by=['AllHomes'],ascending=False).head(5)
Top5States.plot.barh(title='US ZHVI @ November 2022', grid=True, figsize=(14,7), width=0.9, linewidth=0.75, edgecolor=(0.1,0.5,1,1))
```

```python
Top5States['AllHomes']
```

```python
def human_format(num, pos):
    magnitude = 0
    while abs(num) >= 1000:
        magnitude += 1
        num /= 1000.0
    # add more suffixes if you need them
    return '%.1f%s' % (num, ['', 'K', 'M', 'G', 'T', 'P'][magnitude])
```

```python
def nice_axes(ax):
    ax.set_facecolor('.9')
    ax.tick_params(labelsize=8, length=0)
    ax.grid(True, axis='x', color='white')
    ax.set_axisbelow(True)
#     ax.xaxis.set_major_formatter(StrMethodFormatter('${x:,g}K'))
    ax.xaxis.set_major_formatter(FuncFormatter(nz))
    [spine.set_visible(False) for spine in ax.spines.values()]
    
fig, ax = plt.subplots(figsize=(5, 2.5), dpi=144)
colors = plt.cm.Accent(range(len(Top5States)))
y = Top5States.index
width = Top5States['AllHomes'].values
ax.barh(y=y, width=width, color=colors);
nice_axes(ax)
```

## Time series analysis

```python
# Plot accumulated returns
def plot_top_accum(pct_returns, time_years, regions_count):
    pct_returns.iloc[-time_years*12:].cumsum().sort_values(by=pct_returns.index[-1],axis=1,ascending=False).iloc[:,:regions_count].plot(
        grid = True,
        style = "-o",
        figsize = (14,7),
        linewidth = 0.9
    )

plot_top_accum(zhvi_timeseries_pct[['CA']].groupby(level='Metro',axis=1).mean(), 10, 10)
```

```python
# Plot accumulated returns
def plot_top_accum(pct_returns, time_years, regions_count):
    pct_returns.iloc[-time_years*12:].cumsum().sort_values(by=pct_returns.index[-1],axis=1,ascending=False).iloc[:,:regions_count].plot(
        grid = True,
        style = "-o",
        figsize = (14,7),
        linewidth = 0.5
    )

plot_top_accum(zhvi_timeseries_pct[['CA']], 2, 7)
```
